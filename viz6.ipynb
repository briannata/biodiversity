```python
import altair as alt
import pandas as pd
import numpy as np


data = pd.read_excel("LPD2022_public.xlsx")
data.head()
```

```python
classes = data[["Class"]]
species = data[["Class", "Species", "Common_name"]]
```

```python
diff_classes = classes.drop_duplicates(keep="first", inplace=False)
filtered_species = species.drop_duplicates(subset=["Class", "Species"], keep="first", inplace=False)
len(filtered_species)
```

```python
display(diff_classes)
```

```python
display(filtered_species)
```

```python
aves: int = (filtered_species["Class"]=="Aves").sum()
mammalia: int = (filtered_species["Class"]=="Mammalia").sum()
reptilia: int = (filtered_species["Class"]=="Reptilia").sum()
amphibia: int = (filtered_species["Class"]=="Amphibia").sum()

# just fish
actinopteri: int = (filtered_species["Class"]=="Actinopteri").sum()
elasmobranchii: int = (filtered_species["Class"]=="Elasmobranchii").sum()
coelacanthi: int = (filtered_species["Class"]=="Coelacanthi").sum()
petromyzonti: int = (filtered_species["Class"]=="Petromyzonti").sum()
dipneusti: int = (filtered_species["Class"]=="Dipneusti").sum()
holocephali: int = (filtered_species["Class"]=="Holocephali").sum()
myxini: int = (filtered_species["Class"]=="Myxini").sum()

fish: int = actinopteri + elasmobranchii + coelacanthi + petromyzonti + dipneusti + holocephali + myxini
```

```python
source = pd.DataFrame({"category": ["Birds", "Mammals", "Reptiles", "Amphibians", "Fish"],
                       "value": [aves, mammalia, reptilia, amphibia, fish]})

base = alt.Chart(source).encode(
    theta = alt.Theta("value:Q", stack = True),
    color = alt.Color("category:N")
).properties(
    title = "Number of Species Organized by Class"
)

pie = base.mark_arc(outerRadius=120)
text = base.mark_text(radius=140, size=20).encode(text="value:Q")

pie + text
```
